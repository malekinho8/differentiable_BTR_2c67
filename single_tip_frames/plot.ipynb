```julia
using Plots
using StatsPlots
using BSON: @save, @load
```

```julia
frames = [1, 5, 10, 20, 100]
```

```julia
i = 1
@load "original_$(frames[i]).bson" tip images nframe nrealization T P_over_lambdas_array diff_tip diff_surf 
violin(log10.(diff_tip[:, 1] ./ 10.0), diff_tip[:, 2] ./ 10.0, ylim=(-0.1, 6.0), #xaxis=:log, 
       xtickfontsize=12, ytickfontsize=12, legendfontsize=12, grid=true, label="$(frames[i]) frame", 
       framestyle=:box, linewidth=0.0, dpi=150, fmt=:png, alpha=0.3, color=i)
i = 3
@load "original_$(frames[i]).bson" tip images nframe nrealization T P_over_lambdas_array diff_tip diff_surf 
violin!(log10.(diff_tip[:, 1] ./ 10.0), diff_tip[:, 2] ./ 10.0, ylim=(-0.1, 6.0), #xaxis=:log, 
       xtickfontsize=12, ytickfontsize=12, legendfontsize=12, grid=true, label="$(frames[i]) frames", 
       framestyle=:box, linewidth=0.0, dpi=150, fmt=:png, alpha=0.3, color=i)

i = 5
@load "original_$(frames[i]).bson" tip images nframe nrealization T P_over_lambdas_array diff_tip diff_surf 
violin!(log10.(diff_tip[:, 1] ./ 10.0), diff_tip[:, 2] ./ 10.0, ylim=(-0.1, 6.0), #xaxis=:log, 
       xtickfontsize=12, ytickfontsize=12, legendfontsize=12, grid=true, label="$(frames[i]) frames", 
       framestyle=:box, linewidth=0.0, dpi=150, fmt=:png, alpha=0.3, color=i, legend=:bottomleft)

#for i = 2:length(frames)
#    violin!(log10.(diff_tip[:, 1]), diff_tip[:, 2] / 10.0, ylim=(-0.1, 6.0), #xaxis=:log, 
#           xtickfontsize=12, ytickfontsize=12, legendfontsize=12, grid=false, label="$(frames[i]) frames", 
#           framestyle=:box, linewidth=0.0, dpi=150, fmt=:png, alpha=0.3, color=i)
#end

xlabel!("log thresh [nm]", fontsize=35)
p = ylabel!("RMSD from true tip [nm]", fontsize=35)
ylims!(0.0, 7.0)
```
